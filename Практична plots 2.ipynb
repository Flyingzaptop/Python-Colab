# Завдання 3
Очистіть дані від викидів. Оберіть стовпчики на основі ящиків з вусами

```python
q1p = df['Pressure (millibars)'].quantile(0.25)
q3p = df['Pressure (millibars)'].quantile(0.75)
iqr = q3p - q1p
lowerp = q1p - 1.5*iqr
upperp = q3p + 1.5*iqr

q1w = df['Wind Speed (km/h)'].quantile(0.25)
q3w = df['Wind Speed (km/h)'].quantile(0.75)
iqr = q3w - q1w
lowerw = q1w - 1.5*iqr
upperw = q3w + 1.5*iqr

df = df[(df['Pressure (millibars)'] > lowerp) & (df['Pressure (millibars)'] < upperp)]
df = df[(df['Wind Speed (km/h)'] > lowerw) & (df['Wind Speed (km/h)'] < upperw)]
```

```python
df['Pressure (millibars)'].plot.box(figsize=(14, 6),vert=False)
```

```python
df['Wind Speed (km/h)'].plot.box(figsize=(14, 6),vert=False)
```

Тепер Викидні не є критичними
